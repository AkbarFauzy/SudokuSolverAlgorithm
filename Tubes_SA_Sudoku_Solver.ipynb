> Sudoku Solver using Backtracking




```python
import time
```

```python
# Fungsi umtuk mencari sebuah lokasi yang berisikan angka 0
def find_empty_location(arr, l):
    for row in range(9):
        for col in range(9):
            if (arr[row][col] == 0):
                l[0]= row
                l[1]= col
                return True
    return False
 
#Fungsi untuk mengecek apakah nomor sudah ada pada row
def used_in_row(arr, row, num):
    for i in range(9):
        if (arr[row][i] == num):
            return True
    return False
 
#Fungsi untuk mengecek apakah nomor sudah ada pada col 
def used_in_col(arr, col, num):
    for i in range(9):
        if (arr[i][col] == num):
            return True
    return False
 
#Fungsi untuk mengecek apakah nomor sudah ada pada sebuah kotak 3x3 
def used_in_box(arr, row, col, num):
    for i in range(3):
        for j in range(3):
            if (arr[i + row][j + col] == num):
                return True
    return False

#Fungsi mendapatkan semua kemungkinan
def get_all_possible_values(arr,row,col):
  values = [] #Inisialisasi array
  values.append(row) #index 0 sebagai row
  values.append(col) #index 1 sebagai col
  for num in range (1, 10):
    if (not used_in_row(arr, row, num) and not used_in_col(arr, col, num) and not used_in_box(arr, row - row % 3, col - col % 3, num)): #Jika tidak ada pada row/col/box
      values.append(num)
  return values

#Fungsi Print Array
def print_grid(arr):
    for i in range(9):
        for j in range(9):
            print(arr[i][j], end = ' ')
        print('')
```

> Sudoku Solver using Backtracking


```python
def solveSudokuBacktrack(arr):
     
    l =[0, 0]
     
    if (not find_empty_location(arr, l)): # Jika tidak menemukan lokasi dengan angka 0 maka selesai (solusi ditemukan)
        return True
    row = l[0]
    col = l[1]
     
    for num in range(1, 10): #Perulangan dari 1 sampai 9
        #cek apakah value num ada pada sebuh row dan col yang sejajar serta pada sebuah kotak 3x3 
        if (not used_in_row(arr, row, num) and not used_in_col(arr, col, num) and not used_in_box(arr, row - row % 3, col - col % 3, num)): 
             
            arr[row][col]= num 
 
            if (solveSudokuBacktrack(arr)): # Jika success maka lanjut
                return True
            arr[row][col] = 0 # Jika gagal coba angka lainnya
    return False # Backtrack Tringger (jika tidak ada angka yang berhasil)
```



> Sudoku Solver using Branch and Bound



```python
def solveSudokuBnB(arr):
    possibilities = [] #Inisialisasi array
    for i in range(9):  #Perulangan untuk row
      for j in range(9): #perulangan untuk col
        if (arr[i][j] == 0): #Jika kordinat kosong
          possibilities.append(get_all_possible_values(arr,i,j)) #Mendapatkan semua kemungkinan dan meyimpannya ke dalam array
    possibilities.sort(key=len) #Sorting berdasarkan jumlah kemungkinan, python sort menggukanakan algoritma Timshort dengan kompleksitas waktu O(n log (n))
    return solve(arr,0,possibilities) #Algoritma Branch
    
def solve(arr, cpos, possibilities):
    row =  possibilities[cpos][0] #kordinat row
    col =  possibilities[cpos][1] #kordinat col
    for i in range(2,len(possibilities[cpos])): #perulangan dari array index ke dua hingga akhir
      num = possibilities[cpos][i]

      #Jika tidak ada pada row/col/box
      if (not used_in_row(arr, row, num) and not used_in_col(arr, col, num) and not used_in_box(arr, row - row % 3, col - col % 3, num)): 
        arr[row][col] = num
        if (cpos + 1 >= len(possibilities) or solve(arr,cpos + 1,possibilities)): #Jika semua kemungkinan sudah habis atau ada solusi (fungsi rekursif)
          return True
        else: #Jika prune (bukan solusi)
          arr[row][col] = 0 
    return False
```
